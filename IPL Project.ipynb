```python


# IPL dataset load kar le
ipl = pd.read_csv("matches.csv")  # yeh file ka naam tu change kar liyo agar alag hai

# Player of the Match ke counts nikaal lo
player_counts = ipl['player_of_match'].value_counts()

# Top 10 players ke naam aur unke counts nikaal lein (chart clean dikh sake isliye 10 players le rahe hain)
top_players = player_counts.head(10)
players = top_players.index.tolist()       # Player ke naam ki list
counts = top_players.values.tolist()       # Unke "Player of the Match" count ki list

# Pie Chart banana start karein
plt.figure(figsize=(5,4))
plt.pie(counts, labels=players, autopct='%1.1f%%', startangle=140)
plt.title("Top 10 Player of the Match Winners in IPL", fontsize=14)
plt.axis('equal')  # Yeh chart ko gol banata hai (circle)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset
ipl = pd.read_csv("matches.csv")

# Step 2: Select only matches won by batting first
batting_first = ipl[ipl['win_by_runs'] != 0]

# Step 3: Count how many times each team won by batting first
team_wins = batting_first['winner'].value_counts().head(10)

# Step 4: Prepare data for plotting
teams = team_wins.index.tolist()
wins = team_wins.values.tolist()

# Step 5: Plotting bar chart
plt.figure(figsize=(6,4))
plt.bar(teams, wins, color='skyblue')
plt.xlabel("Teams")
plt.ylabel("Wins by Batting First")
plt.title("Top 10 Teams (Batting First Wins)", fontsize=12)
plt.xticks(rotation=45)# for if names are long.
plt.tight_layout()#it's basically a command for pandas to say make a space for big data.
plt.show()
```
